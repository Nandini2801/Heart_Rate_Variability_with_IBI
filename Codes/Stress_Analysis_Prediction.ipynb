***Importing neccessary modules***

```python
import os
import pickle
import pandas as pd
import numpy as  np
```

***Loading the XGboost model***

```python
xgb_clf = pickle.load(open("/content/drive/MyDrive/FYP/first/final/Saved_Model/xgb_clf.pkl", "rb")) #path of the saved model needs to be provided
```

***Generating Predictions from HRV Parameters***

```python
for i in os.listdir('/content/drive/MyDrive/FYP/first/final/Results/HRV_Parameters'): #Path of folder containing HRV Parameters
  df=pd.read_csv("/content/drive/MyDrive/FYP/first/final/Results/HRV_Parameters/{}".format(i))
  test_sample = df.drop(['Unnamed: 0'], axis=1)
  #if SDANN,SDNNI is Present:
  #test_sample=test_sample.drop(['SDANN','SDNNI'],axis=1)

  #test_sample
  y_pred_new = xgb_clf.predict(test_sample)
  y_pred_new = pd.DataFrame(y_pred_new)


  #printing combined parameters and prediction together
  Combined_df= pd.concat([test_sample,y_pred_new],axis=1)
  Combined_df.rename(columns = {0:'Stress_Levels'}, inplace = True)

  # Renaming the 0,1,2 to Stress levels
  Stress_map={0:'no_stress(Low_stress)',1:"time_pressure(Med_stress)",2:"interruption(High_stress)"}
  Combined_df['Stress_Levels']=Combined_df['Stress_Levels'].map(Stress_map)


  #Saving the Prediction Result(Stress Levels) along with the HRV parameters
  Combined_df.to_csv("{}_combined_parameters_pred.csv".format(i.split('.')[0]))
```
